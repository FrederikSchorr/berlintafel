```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import fbprophet
```

### New members

```python
data_newmembers = pd.read_csv("data/newmembers.csv")
```

```python
data_newmembers.tail()
```

```python
data_newmembers.columns = ["personalID", "ds"]
data_newmembers.ds = pd.to_datetime(data_newmembers.ds)
```

```python
members = data_newmembers[["ds"]].copy()
members["delta"] = 1
members.set_index("ds", inplace=True)
```

### Resignations

```python
data_resignations = pd.read_csv("data/resignations.csv")
```

```python
data_resignations.tail()
```

```python
data_resignations.columns = ["personalID", "ds"]
data_resignations.ds = pd.to_datetime(data_resignations.ds)
```

```python
resignations = data_resignations[["ds"]].copy()
resignations["delta"] = -1
resignations.set_index("ds", inplace=True)
```

```python
members_resignations = pd.concat([members, resignations], axis=0)
members_resignations.sort_index(inplace=True)
members_resignations = members_resignations.iloc[1:]
```

```python
members_resignations["y"] = members_resignations.delta.cumsum()
```

```python
members_resignations.head()
```

```python
members_resignations.plot()
```

```python
members_resignations.reset_index(level=0, inplace=True)
del members_resignations["delta"]
```

```python
members_resignations.tail()
```

### Prophet

```python
# Make the prophet model and fit on the data
members_prophet = fbprophet.Prophet()
members_prophet.fit(members_resignations)
```

```python
# Make a future dataframe for 2 years
future = members_prophet.make_future_dataframe(periods=365 * 1, freq='D')
# Make predictions
members_forecast = members_prophet.predict(future)
```

```python
members_prophet.plot(members_forecast, xlabel = 'Date', ylabel = 'New members')
```

```python
members_prophet.plot_components(members_forecast)
```
